```python
import gym
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from torch.distributions import Categorical
from torch.utils.data.sampler import BatchSampler, SubsetRandomSampler
import matplotlib.pyplot as plt
from collections import namedtuple
from IPython.display import clear_output
from statistics import mean
```

```python
class Actor_critic(nn.Module):
    def __init__(self, in_dim, out_dim):
        super(Actor_critic, self).__init__()
        
        self.critic_net = nn.Sequential(
            nn.Linear(in_dim, 64),
            nn.ReLU(),
            nn.Linear(64, 64),
            nn.ReLU(),
            nn.Linear(64, 1)
        )
        
        self.actor_net = nn.Sequential(
            nn.Linear(in_dim, 64),
            nn.ReLU(),
            nn.Linear(64, 64),
            nn.ReLU(),
            nn.Linear(64, out_dim),
            nn.Softmax(dim=1)
        )
    
    def forward(self, x):
        value = self.critic_net(x)
        probs = self.actor_net(x)
        
        return Categorical(probs), value
```

```python
def smooth_plot(factor, item, plot_decay):
    item_x = np.arange(len(item))
    item_smooth = [np.mean(item[i:i+factor]) if i > factor else np.mean(item[0:i+1])
                  for i in range(len(item))]
    for i in range(len(item)// plot_decay):
        item_x = item_x[::2]
        item_smooth = item_smooth[::2]
    return item_x, item_smooth
    
def plot(episode, rewards, losses, episode_steps):
    clear_output(True)
    rewards_x, rewards_smooth = smooth_plot(10, rewards, 500)
    losses_x, losses_smooth = smooth_plot(10, losses, 100000)
    episode_steps_x, episode_steps_smooth = smooth_plot(10, episode_steps, 350)
    
    plt.figure(figsize=(18, 12))
    plt.subplot(311)
    plt.title('Episode %s. reward: %s'%(episode, rewards_smooth[-1]))
    plt.plot(rewards, label="Rewards", color='lightsteelblue', linewidth='1')
    plt.plot(rewards_x, rewards_smooth, label='Smothed_Rewards', color='darkorange', linewidth='3')
    plt.legend(loc='best')
    
    plt.subplot(312)
    plt.title('Losses')
    plt.plot(losses,label="Losses",color='lightsteelblue',linewidth='1')
    plt.plot(losses_x, losses_smooth, 
             label="Smoothed_Losses",color='darkorange',linewidth='3')
    plt.legend(loc='best')
    
    plt.subplot(313)
    plt.title('Episode_Steps %s. Steps: %s'%(episode, episode_steps_smooth[-1]))
    plt.plot(episode_steps,label="Episode_Steps",color='lightsteelblue',linewidth='1')
    plt.plot(episode_steps_x, episode_steps_smooth, 
             label="Episode_Steps_Losses",color='darkorange',linewidth='3')
    plt.legend(loc='best')
    
    plt.show()
```

```python
## hyperparameters

env_name = "CartPole-v0"
max_epoch = 10000
num_timesteps = 300
clip_param = 0.2
ppo_epochs = 10
mini_batch_size = 10
gamma = 0.98
learning_rate = 0.001
max_grad_norm = 0.5

## hyperparameters

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

env = gym.make(env_name)

in_dim = env.observation_space.shape[0]
out_dim = env.env.action_space.n
network = Actor_critic(in_dim, out_dim).to(device)
optimizer = optim.Adam(network.parameters(), lr=learning_rate)

Trajectory = namedtuple('Trajectory', ['state', 'action', 'reward', 'log_prob'])
buffer = []
```

```python
def choose_action(state):
    state_to_network = torch.FloatTensor(state).unsqueeze(0).to(device)
    with torch.no_grad():
        dist, value = network(state_to_network)
    action = dist.sample()
    next_state, reward, done, _ = env.step(action.item())

    buffer.append(Trajectory(state, action.item(), reward, dist.log_prob(action).item()))
    
    return next_state, reward, done
```

```python
def ppo_train():
    losses = []
    
    state = torch.FloatTensor([t.state for t in buffer]).to(device)
    action = torch.LongTensor([t.action for t in buffer]).view(-1, 1).to(device)
    reward = [t.reward for t in buffer]
    old_log_prob = torch.FloatTensor([t.log_prob for t in buffer]).view(-1, 1).to(device)
    
    R, Gt = 0, []
    for r in reversed(reward):
        R = r + gamma * R
        Gt.insert(0, R)
    Gt = torch.FloatTensor(Gt).to(device)
    
    for _ in range(ppo_epochs):
        for index in BatchSampler(SubsetRandomSampler(range(len(buffer))), mini_batch_size, False):
            Gt_index = Gt[index].view(-1, 1)
            dist, value = network(state[index])
            entropy = dist.entropy().mean()
            advantage = (Gt_index - value).detach()
            
            action_prob =  dist.log_prob(action[index])
            
            ratio = (action_prob - old_log_prob[index]).exp()
            surr1 = ratio * advantage
            surr2 = torch.clamp(ratio, 1 - clip_param, 1 + clip_param) * advantage
            actor_loss = - torch.min(surr1, surr2).mean()
            
            critic_loss = (Gt_index - value).pow(2).mean()
            
            loss = 0.5 * critic_loss + actor_loss - 0.01 * entropy
            losses.append(loss.item())
            
            optimizer.zero_grad()
            loss.backward()
            nn.utils.clip_grad_norm_(network.parameters(), max_grad_norm)
            optimizer.step()
    
    del buffer[:]
    return round(mean(losses), 2)
```

batch_size设置过大，会导致done掉了，但是和环境互动的step不够，不能做到持续学习，持续更新，而等到更新时，经验过于陈旧，导致新旧log_prob相差较大，无法正常学习

```python
all_rewards, all_loss, episode_steps = [], [], []

for epoch in range(max_epoch):
    
    state = env.reset()
    rewards = 0
    
    for i in range(num_timesteps):
        
        next_state, reward, done = choose_action(state)
        
        state = next_state
        rewards += reward
        
        if done or i==(num_timesteps-1):
            if len(buffer) >= mini_batch_size:
                loss = ppo_train()
                all_loss.append(loss)
            episode_steps.append(i)
            break
    
    all_rewards.append(rewards)
    plot(epoch, all_rewards, all_loss, episode_steps)
```
